```python
import numpy as np
import cv2 as cv
import math 

rostro = cv.CascadeClassifier('C:/Users/zerog/Documents/Noveno Semestre/IA/source/haarcascade_frontalface_alt.xml')
cap = cv.VideoCapture(0)
while True:
    ret, frame = cap.read()
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    rostros = rostro.detectMultiScale(gray, 1.3, 5)
    for(x, y, w, h) in rostros:
        frame = cv.rectangle(frame, (x,y), (x+w, y+h), (0, 255, 0), 3)
        #frame = cv.rectangle(frame, (x+int(w/5),y+int(h/5)), (x+w-int(w/6), y+h-int(h/2)), (255, 0, 0), 3)#ojos
        frame = cv.circle(frame, (x+int(w/3.3),y+int(h/2.6)), int(w/13), (255, 0, 0), 2)#ojoIzq
        frame = cv.circle(frame, (x+w-int(w/3.3),y+int(h/2.6)), int(w/13), (255, 0, 0), 2)#ojoDer
        frame = cv.rectangle(frame, (x+int(w/2)-int(w/7.9),y+int(h/2)-int(h/6.5)), (x+int(w/2)+int(w/7.9),y+int(h/2)+int(h/6)), (0, 0, 255), 2)#nariz
        frame = cv.rectangle(frame, (x+int(w/2)-int(w/5.5),y+int(h/2)+int(h/3.9)), (x+int(w/2)+int(w/5.5),y+int(h/2)+int(h/2.4)), (0, 250, 250), 2)#boca

        print(str(w)+" "+str(h))
        
        
        
    cv.imshow('rostros', frame)
    k = cv.waitKey(1)
    if k == 27:
        break
cap.release()
cv.destroyAllWindows()

```
