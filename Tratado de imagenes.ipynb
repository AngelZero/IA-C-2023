```python
import cv2 as cv
import numpy as np
```

```python
#05/09/23 Capturar Color En Video 
cap = cv.VideoCapture(0)

if not cap.isOpened():
    print("No se puede aperturar la camara")
    exit()
    
while True:
    ret, img = cap.read()
    
    img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB) #cambia modelo de color
    img3 = cv.cvtColor(img2, cv.COLOR_RGB2HSV)
    
    #Para rojos
    umbralBajo=(0,80,80)
    umbralAlto=(10,255,255)

    umbralBajo2=(170,120,120)
    umbralAlto2=(180,255,255)
    
    #Para azules
    umbralBajoAz=(85,80,80)
    umbralAltoAz=(135,255,255)

    mascara1=cv.inRange(img3,umbralBajo,umbralAlto)#Solamente Tomar los elementos de la imagen que esten comprendidos entre el rango de color que acabo de probar.
    mascara2=cv.inRange(img3,umbralBajo2,umbralAlto2)#Solamente Tomar los elementos de la imagen que esten comprendidos entre el rango de color que acabo de probar.
    mascaraAz=cv.inRange(img3,umbralBajoAz,umbralAltoAz)
    mascaraDef=mascaraAz+mascara2#mascara1+mascara2

    resultado = cv.bitwise_and(img, img, mask=mascaraDef)

    cv.imshow('resultado',resultado)
    cv.imshow('mascara1',mascara1)
    cv.imshow('mascara2',mascara2)
    cv.imshow('BGR',img)
    cv.imshow('RGB',img2)
    cv.imshow('HSV',img3)
    
    #if not ret:
    #    print("No se pueden capturar fotogramas")
    #    break
    
    cv.imshow('frame',img)
    if cv.waitKey(1)==ord('q'):
        break
cap.release()
cv.destroyAllWindows()
```

```python
img = cv.imread('C:/Users/zerog/Documents/Noveno Semestre/IA/hidden.jpg',0)
# Función para ampliar y mostrar un píxel en un área de 100x100 píxeles
def ampliar_pixel(event, x, y, flags, param):
    if event == cv.EVENT_LBUTTONDOWN:  # Detecta un clic izquierdo
        tamano_pixel = 30
        n_matriz = 20
        n = tamano_pixel*n_matriz
        
        newx = int(x - (n_matriz-1)/2)
        newy = int(y - (n_matriz-1)/2)
        
        # Crea una imagen en blanco del tamaño deseado (100x100 píxeles) y establece el valor del píxel
        zoomed_img = np.zeros((n, n, 3), dtype=np.uint8)

        for i in range(n_matriz):
            for j in range(n_matriz):
                # Extrae el valor RGB del píxel clicado
                pixel = img[newy+i, newx+j]
                zoomed_img[i*tamano_pixel:(i+1)*tamano_pixel , j*tamano_pixel:(j+1)*tamano_pixel] = pixel        

        # Muestra el píxel ampliado en una ventana separada
        cv.imshow('Píxel Ampliado', zoomed_img)
        cv.waitKey(0)
        cv.destroyWindow('Píxel Ampliado')  # Cierra la ventana emergente

# Crea una ventana para la imagen
cv.namedWindow('Imagen')
cv.imshow('Imagen', img)

# Establece la función de detección de clics
cv.setMouseCallback('Imagen', ampliar_pixel)

# Muestra la imagen y espera hasta que se presione la tecla 'Esc' para salir
while True:
    cv.imshow('Imagen', img)
    key = cv.waitKey(1)
    if key == 27:  # Presionar la tecla 'Esc' para salir
        break

cv.destroyAllWindows()
```

```python
img = cv.imread('C:/Users/zerog/Documents/Noveno Semestre/IA/hidden.jpg',0)
#cv.imshow('Ejemplo', img)
w,h = img.shape
cv.imshow('EjemploAntesFiltro',img)
img2 = img.copy()

#Binarizar
for i in range(w):
    for j in range(h):
        if(img[i,j]>150):
            img2[i,j]=255
        else:
            img2[i,j]=0
cv.imshow('FiltroBinarizado',img2)

#Negativo
for i in range(w):
    for j in range(h):
            img2[i,j]=255 - img[i,j]
cv.imshow('FiltroNegativo',img2)

#Eliminar canal R
imgc = cv.imread('C:/Users/zerog/Documents/Noveno Semestre/IA/hidden.jpg',1) #1 son los 3 canales
img2 = imgc.copy()
for i in range(w):
    for j in range(h):
        img2[i,j,2]= 0
cv.imshow('SinCanalR',img2)

zero = np.zeros(img.shape[:2], dtype='uint8')#con numeros enteros sin signo de 8 bits
img2 = cv.cvtColor(imgc, cv.COLOR_BGR2RGB) #cambia modelo de color
(b,g,r) = cv.split(imgc)
cv.imshow('b',b)
cv.imshow('g',g)
cv.imshow('r',r)
cv.imshow('b1',cv.merge([b,zero,zero]))
cv.imshow('g1',cv.merge([zero,g,zero]))
cv.imshow('r1',cv.merge([zero,zero,r]))
cv.imshow('BGR2RGB',img2)
cv.imshow('zero',zero)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Función para ampliar y mostrar un área de 600x600 píxeles con el píxel clicado en el centro
def ampliar_vecinos(event, x, y, flags, param):
    if event == cv.EVENT_LBUTTONDOWN:  # Detecta un clic izquierdo
        # Define el tamaño del área ampliada (600x600 píxeles)
        area_size = 600

        # Calcula las coordenadas del área ampliada
        x1, y1 = max(x - area_size // 2, 0), max(y - area_size // 2, 0)
        x2, y2 = min(x + area_size // 2, img.shape[1]), min(y + area_size // 2, img.shape[0])

        # Extrae el área ampliada
        zoomed_area = img[y1:y2, x1:x2].copy()

        # Crea una imagen en blanco del tamaño deseado (600x600 píxeles)
        zoomed_img = np.zeros((area_size, area_size), dtype=np.uint8)
        
        # Calcula las coordenadas del píxel clicado en el área ampliada
        center_x, center_y = area_size // 2, area_size // 2

        # Copia los píxeles de la zona ampliada alrededor del píxel clicado en la imagen ampliada
        zoomed_img[
            center_y - (y2 - y1) // 2: center_y + (y2 - y1) // 2,
            center_x - (x2 - x1) // 2: center_x + (x2 - x1) // 2
        ] = zoomed_area

        # Muestra el área ampliada en una ventana separada
        cv.imshow('Área Ampliada', zoomed_img)
        cv.waitKey(0)
        cv.destroyWindow('Área Ampliada')  # Cierra la ventana emergente

# Crea una ventana para la imagen
cv.namedWindow('Imagen')
cv.imshow('Imagen', img)

# Establece la función de detección de clics
cv.setMouseCallback('Imagen', ampliar_vecinos)

# Muestra la imagen y espera hasta que se presione la tecla 'Esc' para salir
while True:
    cv.imshow('Imagen', img)
    key = cv.waitKey(1)
    if key == 27:  # Presionar la tecla 'Esc' para salir
        break

cv.destroyAllWindows()
```
